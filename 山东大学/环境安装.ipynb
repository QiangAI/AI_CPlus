## 1. 安装Qt

1. 使用安装文件
    - Qt的源代码安装：qt-everywhere-src-5.14.2.zip

2. 解压缩
    - 直接使用文件资源管理器解压缩即可：

- 解压缩后的截图：
    

3. 安装前的编译环境
    - 使用vcvars64.bat启动Visual Studio的编译环境。
    - 执行命令： `vcvars64.bat`

3. 生成编译工程
    - 编译指令：`configure.bat -prefix "C:\Qt514" -confirm-license -opensource  -debug-and-release -platform win32-msvc`
        - `-prefix` : 指定安装目录，会自动创建；
        - `-confirm-license` : 确认同意许可；
        - `-debug-and-release` : 编译安装Debug与Release版本
        - `-platform` : 指定平台编译器，win32-msvc就是Visual Stduio

4. 编译
    - 编译指令： `nmake` 

5. 安装：
    - 安装指令：`nmake install`

## 安装OpenCV

1. 文件准备
    - opencv4.2.0.zip
    - opencv4.2.0-contrib.zip  

2. 解压缩
    - 使用文件资源管理器解压缩，并把opencv4.2.0-contrib的文件拷贝到opencv4.2.0目录中实现目录合并。

3. 修改几个配置文件，把`https://raw.githubusercontent.com`替换为`https://raw.staticdn.net`：
    - 3rdparty
        - ippicv
        - ffmpeg
    - modules
        - face
        - xfeatures2d/cmake

- 注意：
    - 详细的替换请参考：
        - `https://raw.staticdn.net`
        - `https://7ed.net/gra/`

4. 使用cmake gui配置
    - 因为不使用GPU，所以采用默认配置；
    - 其中会出现一些错误，根据情况修改，但是有个错误：`CONFIGURATION IS NOT SUPPORTED: validate setupvars script in install`,直接取消这个功能即可。
